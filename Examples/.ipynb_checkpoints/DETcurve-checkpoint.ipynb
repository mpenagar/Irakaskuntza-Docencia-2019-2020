```python
from matplotlib import pyplot as plt
from matplotlib import ticker
def DETCurve(fps,fns):
    """
    Given false positive and false negative rates, produce a DET Curve.
    The false positive rate is assumed to be increasing while the false
    negative rate is assumed to be decreasing.
    """
    axis_min = min(fps[0],fns[-1])
    fig,ax = plt.subplots()
    plt.plot(fps,fns)
    plt.yscale('log')
    plt.xscale('log')
    ticks_to_use = [0.001,0.002,0.005,0.01,0.02,0.05,0.1,0.2,0.5,1,2,5,10,20,50]
    ax.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
    ax.get_yaxis().set_major_formatter(ticker.ScalarFormatter())
    ax.set_xticks(ticks_to_use)
    ax.set_yticks(ticks_to_use)
    plt.axis([0.001,50,0.001,50])
```

```python
fps = [ (1-x/100) for x in range(100) ]
fns = [ 1-x for x in fps]
#print(fps)
#print(fns)
DETCurve(fps,fns)
```

```python
import scipy.stats
tar = scipy.stats.norm(2, 1)
non = scipy.stats.norm(-4, 1)
n = 100
xmax = 5
X = [(x/(n-1)*2*xmax)-xmax for x in range(n)]
fps = [(1.0-non.cdf(x)) * 100  for x in X]
fns = [tar.cdf(x) * 100 for x in X]
DETCurve(fps,fns)
```

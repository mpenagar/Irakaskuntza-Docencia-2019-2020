```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
mu, sigma = 100, 15
samples = 10000
#x = mu + sigma * np.random.randn(10000)
x = np.random.normal(mu, sigma, samples)
```

```python
# plots the histogram of the data and returns xlim,ylim
def plothist(*args,legend=None,xlim=None,ylim=None):
    for data in args :
        plt.hist(data, bins='auto', density=False)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.grid(True)
    if xlim : plt.xlim(*xlim)
    if ylim : plt.ylim(*ylim)
    xlim,ylim = plt.xlim(),plt.ylim()
    if legend :
        xpos = xlim[0] + 0.10 * (xlim[1]-xlim[0])
        ypos = ylim[0] + 0.85 * (ylim[1]-ylim[0])
        plt.text(xpos,ypos,legend)
    plt.show()
    return xlim,ylim
```

```python
xlim,ylim = plothist(x,legend="$\mu={},\ \sigma={}$".format(mu, sigma))
```

```python
threashold = 115
x2 = [ i for i in x if i > threashold]
samples2 = len(x2)
legend = "{}% of samples".format(int(samples2/samples*100))
xlim,ylim = plothist(x2,legend=legend,xlim=xlim,ylim=ylim)
```

```python
def createValues(mu,sigma,N,n):
    N += 1
    return [ mu + sigma * norm.ppf(i/N) for i in range(1,n+1) ]
for N,n in ((100,100),(100,50),(100,10)) :
    v = createValues(0,1,N,n)
    print(len(v),min(v),max(v))
```
